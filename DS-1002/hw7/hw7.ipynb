## 4b

```python
states = [1,2,3]
```

```python
mu0 = np.array([1,0,0])
```

```python
def simulate(mu0, P, n=10000):
    states = list(range(1, len(mu0)+1))
    current = np.random.choice(states, p=mu0)
    path = [current]
    for i in range(n):
        # Use row of transition matrix to choose next state
        current = np.random.choice(states, p=P[current-1,:])
        path.append(current)      
    return np.array(path)    
```

```python
sample = simulate(mu0, mat)
```

```python
plt.plot(sample[:25])
plt.xlabel('t')
plt.ylabel('State')
```

## 4c

```python
def get_trajectory(sample, state):
    trajectory = []
    for i in range(len(sample)):
        subsample = sample[:i]
        freq = len(subsample[subsample == state])/(i+1)
        trajectory.append(freq)
    return trajectory
        
```

```python
for state in states:
    plt.plot(get_trajectory(sample, state), label='State {}'.format(state))
plt.legend(loc='right')
plt.xlabel('t')
plt.ylabel('Frequency of state')
```

State 1 occurs about 10% of the time, state 3 about 30% of the time, and state 2 around 60% of the time.
